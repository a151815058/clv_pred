```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
customerOrderDT = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/佐登妮絲/青年計劃/clv_pred/customerOrderDT_.csv')
customerOrderDT
```

```python
allDatesDT = pd.DataFrame(customerOrderDT['InvoiceDate'].unique(),columns={'InvoiceDate'}).sort_index()

allDatesDT['InvoiceDate'] = pd.to_datetime(allDatesDT['InvoiceDate'])
```

```python
allDatesDT['DayID'] = (allDatesDT['InvoiceDate'] - allDatesDT['InvoiceDate'].min()).dt.days

allDatesDT
```

#### 轉換成每個CustomerID每一天購買的金額與item數量

```python
customerOrderDayDT_ItemProfit = customerOrderDT.groupby(['CustomerID','InvoiceDate']).sum('Profit')['Profit'].reset_index()
customerOrderDayDT_UniqueItems = customerOrderDT.groupby(['CustomerID','InvoiceDate'])['StockCode'].nunique().reset_index()
customerOrderDayDT = pd.merge(customerOrderDayDT_ItemProfit,customerOrderDayDT_UniqueItems)
customerOrderDayDT
```

```python
customerOrderDayDT.info()
```

```python
customerOrderDayDT['InvoiceDate'] = pd.to_datetime(customerOrderDayDT['InvoiceDate'])
```

```python
customerOrderDayDT = pd.merge(customerOrderDayDT,allDatesDT, on ='InvoiceDate')
```

```python
customerOrderDayDT
```

```python
customerOrderDayDT['timeID'] = customerOrderDayDT['DayID']
customerOrderTimeDT_ItemProfit = customerOrderDayDT.groupby(['CustomerID','timeID']).sum('ItemProfit')['Profit'].reset_index()
customerOrderTimeDT_UniqueItems = customerOrderDayDT.groupby(['CustomerID','timeID']).sum('StockCode')['StockCode'].reset_index()
customerOrderTimeDT = pd.merge(customerOrderTimeDT_ItemProfit,customerOrderTimeDT_UniqueItems)
customerOrderTimeDT
```

```python
customerOrderTimeDT_CustomerID = pd.DataFrame(customerOrderTimeDT['CustomerID'].unique())
customerOrderTimeDT_timeID = pd.DataFrame(customerOrderTimeDT['timeID'].unique())

allCustTimeDT = pd.merge(customerOrderTimeDT_CustomerID,customerOrderTimeDT_timeID,how='outer')
allCustTimeDT
```
